```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import geopandas as gpd
import seaborn as sns
```

# Data Exploartion and cleaning

```python
Data=pd.read_csv('2025-02-west-midlands-street.csv')
Data
```

```python
print(Data.info())
print(Data.describe())
```

```python
# Handling NaN values

print(Data.isna().sum()) #check the sum of null values
# Check if any rows have no missing values
print("Rows with no missing values:", len(Data.dropna()))
#drop Nan column
dropped_column=Data.drop(columns=['Context']) 
#Removing nan values
clean_data=dropped_column.dropna(subset=['Crime ID','Last outcome category'])
clean_data.info()
```

```python
#handling datatypes 
clean_data['Month']=pd.to_datetime(clean_data['Month']) #changing month to date format
```

# Descriptive statistics 

```python
crime_type=Data['Crime type'].value_counts()
print(f'The crime counts are:\n {crime_type} \n')
print (f' The crime type mean is: \n {crime_type.mean()}')
```

```python
# Crime frequency by type
crime_type.plot(kind='bar', color='blue')
#visualizing the crime frequency
plt.title('Crime frequency by type')
plt.xlabel('crime type')
plt.ylabel('count')
plt.show()
```

```python
# Count crimes per LSOA
crime_by_lsoa = clean_data.groupby('LSOA name').size().reset_index(name='crime_count')
print(crime_by_lsoa)


# Visualize top 5 LSOAs
import matplotlib.pyplot as plt
top_lsoas = crime_by_lsoa.sort_values('crime_count', ascending=False).head(10)
plt.bar(top_lsoas['LSOA name'], top_lsoas['crime_count'])
plt.title('Top 10 LSOAs by Crime Count')
plt.xlabel('LSOA Name')
plt.ylabel('Crime Count')
plt.xticks(rotation=45)
plt.show()
```

# Spatial Analysis

```python
import folium
from folium.plugins import HeatMap

# Creating a GeoDataFrame using longitude and latitude 
gdf = gpd.GeoDataFrame(clean_data, geometry=gpd.points_from_xy(clean_data['Longitude'], clean_data['Latitude']))

# Create a heatmap
m = folium.Map(location=[clean_data['Latitude'].mean(), clean_data['Longitude'].mean()], zoom_start=12)
heat_data = [[row['Latitude'], row['Longitude']] for index, row in clean_data.iterrows()]
HeatMap(heat_data).add_to(m)
m.save('crime_heatmap.html')  # Save map as HTML

```

```python
# Crosstab for categorical variables
crime_location = pd.crosstab(clean_data['Crime type'].head(), clean_data['Location'])
sns.heatmap(crime_location, cmap='Blues')
plt.title('Crime Types by Location')
plt.show()
```
